九九乘法

```python
for i in range(2,10):
  for j in range(1,10):
    product = i * j
    print("%d*%d=%2d" %(i,j,product),end = " ") #end:寫完一整行不換行
  print()
```

最小公倍數

```python
a = int(input("請輸入a的值:"))
b = int(input("請輸入b的值:"))

maxno = a * b
for i in range(1,maxno+1):
  if (i % a == 0 and i %b == 0):
    break
print("%d和%d的最小公倍數 = %d" %(a,b,i))
```

排除5的倍數

```python
n = int(input("請輸入一個正整數:"))
for i in range (1,n+1):
  if i % 5 == 0:
    continue
  print(i,end = " ")
```

while迴圈應用

```python
total = i = 1
n = int(input("請輸入一個正整數n的值:"))
while i <= n:
  total *= i 
  i += 1
print("%d!=%d" %(n,total))
```

```python
total1 = n = 0
while(n <= 50):
  total1 += n
  n += 2
print("0~50的偶數和結果是",total1)

total2 = 0
for i in range(0,50):
  if (i %2 == 1):
    total2 += i
print("0~50的奇數和結果是",total2)
```
